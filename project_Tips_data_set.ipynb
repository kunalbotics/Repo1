```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head(10)
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="smoker", size="size",)
plt.title("Scatter Plot of Total Bill vs. Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.barplot(x="smoker", y="total_bill", data=tips,color="red")


plt.title("Average Total Bill for Smokers and Non-Smokers")
plt.xlabel("Smoker")
plt.ylabel("Average Total Bill")


plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x="time", y="total_bill", data=tips)
plt.title("Total Bill Distribution for Different Meal Times")
plt.xlabel("Meal Time")
plt.ylabel("Total Bill")

plt.show()
```

```python
import matplotlib.pyplot as plt


smoker_counts = tips['smoker'].value_counts()


plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%')


plt.title("Proportion of Smokers and Non-Smokers")


plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.violinplot(x="tip", y="day", data=tips)


plt.title("Distribution of Total Bill for Smokers and Non-Smokers")
plt.xlabel("tip")
plt.ylabel("Total Bill")


plt.show()
```
